```python
# Short Theory to know.
# Dictionaries are used to store data values in key:value pairs.
# A dictionary is a collction which is ordered, changeable, and do not allow duplicates.
```

```python
# Creation of Dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print(thisdict)
```

```python
# Retrieving values from dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print(thisdict["brand"])
```

```python
# Note: Dictionary takes the latest value given for key. It will not allow duplicates.
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964,
    "year": 2020
}
print(thisdict)
```

```python
# Length of key-value pairs
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
len(thisdict)
```

```python
# Changing the key:value pair in dictionary
car = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

car["colour"] = "red"
print(car)
x = car.values()
print(x)
print(type(car))
```

```python
# Retrieval of Values
student = {
    "Roll Number": 88,
    "Name": "Rachit",
    "branch": "ENTC"
}
print(student.values())
```

```python
# Removal of a key value pair
student = {
    "Roll Number": 88,
    "Name": "Rachit",
    "branch": "ENTC"
}
student.pop("branch")
print("After removal: ", student)
```

```python
# DIY Problem 2
# A dictionary stores student names and marks.
# Write a python program to search for student's marks.
# If the student is not found, display a suitable message.

marks = {
    "Maahi": 85,
    "Rachit": 92,
    "Madhur": 78,
    "Pranav": 89
}

name = input("Enter a student's name: ")
print(marks.get(name, "Student not found"))
```

```python
# DIY Problem 3
# A dictionary stores usernames and passowrds.
# Write a python program to validate user login.

users = {"admin": '1234', "student": "abcd"}

username = input("Enter username: ")
password = input("Enter password: ")

stored_password = users.get(username)

if stored_password == password:
    print("Login successful")
else:
    print("Invalid username or password")
```
