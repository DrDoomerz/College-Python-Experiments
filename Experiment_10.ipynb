```python
# Code 2: Creating Dataframes
# One column, then use series... Multiple columns, then use dataframes
data = {
  "Name":["A","B","C"],
  "Marks":[85,90,78]
}
DF = pd.DataFrame(data)
print(DF)
```

```python
# Just some short theory: df.head() and df.tails can be used to give the first five rows and the last five rows respectively... from a given table of data.
```

```python
# Command 1: Finding the structure of dataframes
print(DF.shape)
```

```python
# Command 2: Finding the dimension of the data frame
print(DF.ndim)
```

```python
# Command 3: Finding the size of the dataframe
print(DF.size)
```

```python
# Command 4: Finding the columns in the dataframe
print(DF.columns)
```

```python
# Command 5: Checking the datatypes in each column of the dataframe
print(DF.dtypes)
```

```python
# Command 6: Accessing the individual column in the dataframe -
print(DF["Name"])
```

```python
# Command 7: Accessing the element in an individual row and column. (Method 1)
print(DF.loc[0,"Name"])
```

```python
# Command 8: Accessing the elemt in an individual row and column. (Method 2)
print(DF.iloc[2,1])
```

```python
# Command 9: Adding a new column in the dataframe

DF["Grade"] = ["First Class", "Distinction", "Second Class"]
print(DF)
```

```python
# Code 3: Updating the data in dataframe (Method 1)
# Change marks of student A from 85 to 88.

DF.loc[0,"Marks"] = 88
print(DF)
```

```python
# Code 3: Updating the data in dataframe (Method 2)
# Change marks of student A from 85 to 88
DF.iloc[0,1] = 88
print(DF)
```

```python
# Code 4: Deleting a column from the dataframe
# Command 10 - df.drop
# Note: Here inplace was used because we are overwriting the data.

# Ensure the 'Grade' column exists for the drop operations
DF["Grade"] = ["First Class", "Distinction", "Second Class"]

print("DF before drop operations:")
print(DF)
print()

# Drop 'Grade' by creating a new DataFrame (DF1)
DF1 = DF.drop("Grade", axis=1)
print("DF1 after dropping 'Grade' (new DataFrame):")
print(DF1)
print()

# Drop 'Grade' in-place from the original DataFrame (DF)
DF.drop("Grade", axis=1, inplace=True)
print("DF after dropping 'Grade' in-place:")
print(DF)
```

```python
# Code 5: Performing Basic Statistical Analysis
# Command 11 - df.mean()
# Command 12 - df.min()
# Command 13 - df.max()

mean1 = DF["Marks"].mean()
print("The mean value of the dataframe is: ", mean1)
min1 = DF["Marks"].min()
print("The minimum value in the dataframe is: ", min1)
max1 = DF["Marks"].max()
print("The maximum value in the dataframe is: ", max1)
```

```python
# Code 6: Displaying the students scoring more than 80 only from the dataframe
# This is called filtering!
DF[DF["Marks"] > 80]
```
