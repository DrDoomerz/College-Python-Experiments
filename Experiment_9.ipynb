```python
# DIY 1: Learning aboutn in-built functions using Numpy
# Note: The default datatype is a float.
# If you dont want the period or a float then you have to typecast it into an integer or some other type.

# np.zeros will print 0 in the defined number of rows and columns.
print(np.zeros((2,3)))
print()

# np.ones will print 1 in the defined number of rows and columns.
print(np.ones((3,3)))
print()

# np.eye will print a diagonal matrix in the defined number.
print(np.eye(3))
print()

# np.arange will print the values in ascending order based on the limits and spacing.
print(np.arange(1,10,2))
print()

# np.linspace() generates evenly spaced numbers between a specified start and end value (including both endpoints) in a given number of steps.
print(np.linspace(0,1,4))
print()
# linspace formula is: Step Size = (stop - start) / (num - 1)

print(b * 2)
print()
# print(c) - consider it a new matrix c
print(a + 5)
print()
# print(d) - cibsuder it a new matrix d

# np.mean command prints the mean values of the matrix defined.
print(np.mean)
print()
print(np.mean(a))
print()
print(np.mean(b))
print()

# np.median command prints the median values of the matrix defined.
print(np.median(a))
print()
print(np.median(b))
print()

# np.max and np.min command prints the maximum and the minimum value of the matrix.
print(np.max(a))
print()
print(np.min(a))
print()

# `print(np.sum(a))` → Prints the sum of all elements in array `a`.
print(np.sum(a))
print()



# print(np.std(a)) → Prints the standard deviation of elements in array `a`.
print(np.std(a))
print()

# print(np.var(a)) → Prints the variance of elements in array `a`.
print(np.var(a))
print()
```

```python
# DIY 2: Take a matrix and perform all operations on it.
# Note: I was trying to implement this code on my own. It was not discussed in class.
# Numpy can be used in various applications and this is one of them.

import numpy as np

# Define a matrix
A = np.array([[1, 2],
              [3, 4]])

print("Matrix A: ")
print(A)

# Transpose
print("\nTranspose of A: ")
print(A.T)

# Determinant
print("\nDeterminant of A: ")
print(np.linalg.det(A))

# Inverse
print("\nInverse of A: ")
print(np.linalg.inv(A))

# Matrix Multiplication (A × A)
print("\nMultiplication of A with A: ")
print(np.dot(A, A))
```
