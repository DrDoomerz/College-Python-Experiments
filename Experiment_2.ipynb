```python
# Introduction to lists in python
list1 = ["apple", "banana", "orange"]
print(list1)
```

```python
# Writing a list to print an alphanumeric list
list2 = [2,2.3,"rachit", True]
print(list2)
```

```python
# Writing a program to implement the duplication of list
list3 = ["apple", "banana", "orange", "apple", "cherry"]
print(list3)
```

```python
# Writing a program to implement the indexing of lists
list4 = ["Apple", "Banana", "Cherry", "Date", "Elderberry", "Fig", "Grape","Honeydew", "Imbe", "Jackfruit", "Kiwi", "Lemon", "Mango", "Nectarine","Orange", "Papaya", "Quince", "Raspberry", "Strawberry", "Tangerine", "Ugni", "Vanilla Bean", "Watermelon", "Xigua", "Yellow Passion Fruit", "Zucchini (Botanically a fruit)"]
print(list4[2])
print(list4[-4])
print(list4[2:5])
print(list4[:4])
print(list4[2:])
```

```python
# Writing a program to replace an element in the list
list5 = ["apple", "orange", "mango"]
list5[2] = "cherry"
print(list5)
```

```python
# Writing a program to add an element in the list
list6 = ["apple", "orange", "mango"]
list6.append("cherry")
print(list6)
```

```python
# Writing a program to add an element at a specific location in the list
list7 = ["apple", "orange", "mango"]
list7.insert(2, "cherry")
print(list7)
```

```python
# Writing a program to combine multiple lists into one. (part 1)
list8 = ["apple", "banana", "orange"]
list9 = ["kiwi", "mango"]
combined_list = list8 + list9
print(combined_list)
```

```python
# Writing a program to combine multiple lists into one. (part 2)
list8 = ["apple", "banana", "orange"]
list9 = ["kiwi", "mango"]
list8.extend(list9)
print(list8)
print(list9)
```

```python
# DIY Question 1: Write a program to find the maximum marks, minimum marks, total number of students, sum of marks, average marks, sort marks in ascending order.
import math
marks = [90,82,66,76,33]

print("Maximum marks: ", max(marks))
print("Minimum marks: ", min(marks))
print("Total number of students: ", len(marks))
print("Sum of marks: ", sum(marks))
print("Average marks: ", sum(marks)/len(marks))
marks.sort()
print("Marks in ascending order: ", marks)
```

```python
# DIY Question 2
# Scenario: A user maintains a grocery list for monthly shopping.
# Tasks: Create a grocery list, Add new items to the list, Remove purchased items, Display the final shopping list
# self note: try to use append(), remove(), etc. type commands.

grocery_list = ["Milk", "Bread", "Eggs", "Apples", "Chicken"]
print("Initial Grocery List: ", grocery_list)

grocery_list.append("Oranges")
grocery_list.append("Rice")
print("Grocery List after adding items: ", grocery_list)

grocery_list.remove("Eggs")
grocery_list.remove("Bread")
print("Grocery List after removing purchased items: ", grocery_list)

print("Final Shopping List: ", grocery_list)
```
