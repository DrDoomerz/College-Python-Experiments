```python
# Learning how to perform arithmetic operations on variables
a,b = 10, 5
add = a + b
sub = a - b
mul = a * b
div = a / b
mod = a % b
exp = a**b
print(add, sub, mul, div, mod, exp, sep="\n")
print("The addition of the two numbers is : ", add)
print("The substraction of the two numbers is : ", sub)
print("The multiplication of the two numbers is : ", mul)
print("The division of the two numbers is : ", div)
print("The modulus of the two numbers is : ", mod)
print("The exponential value of the two numbers is : ", exp)
```

```python
# Learning to declare a user defined variable for performing arithmetic operations
a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

add = a + b
sub = a - b
mul = a * b
div = a / b
mod = a % b
exp = a**b

print("The addition of the two numbers is : ", add)
print("The subtraction of the two numbers is : ", sub)
print("The multiplication of the two numbers is : ", mul)
print("The division of the two numbers is : ", div)
print("The modulus of the two numbers is : ", mod)
print("The exponential value of the two numbers is : ", exp)

#printing all the results in one line this time...
print("the addition, substraction, multiplication, division, modulus, and exponential value of the two numbers is : ", add, sub, mul, div, mod, exp, sep = "\n")
```

```python
# Writing a code to find the area of a circle... what are the steps?

# Step 1: use the math library
import math

# step 2: Get the radius of the circle from the user
radius = float(input("Enter the radius of the circle: "))

# Step 3: Calculate the area of the circle
area = math.pi * (radius ** 2)

# Step 4: Print the result
print(area)
```

```python
# Writing a code to find the value of a log.

import math

number = float(input("Enter a number to find its log (base 10): "))
log_value = math.log10(number)
print(log_value)
```

```python
# Writing a code to identify which variable is greater in numerical value.
# Note: Make sure to use if and else statement in your code.
# The numbers are user defined here.

a = float(input("Enter the first number: "))
b = float(input("Enter the second number: "))

if a > b:
    print("a is greater than b")
elif b > a:
    print("b is greater than a")
else:
    print("a and b are equal")
```
