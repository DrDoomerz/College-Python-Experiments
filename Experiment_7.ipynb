```python
# DIY 1: Write a program to find the factorial of a number using while loop.

num = int(input("Enter a non-negative integer: "))
factorial = 1
i = 1
# SELF NOTE for future reference:
# The while loop calculates factorial for num >= 0.
# For num < 0, factorial will remain 1, which is not the mathematical definition.
while i <= num:
    factorial *= i
    i += 1
print("The factorial of the given number is: ", factorial)
```

```python
# DIY 2: Write a program to find the fibonacci series using while loop

# for nterms <= 0, as it solely relies on the while loop condition.
# For nterms = 1, it will print only the first term (0).

nterms = int(input("How many terms do you want? "))
n1, n2 = 0, 1
count = 0

print("Fibonacci sequence: ")
while count < nterms:
    print(n1)
    nth = n1 + n2
    # Here we are updating the values now.
    n1 = n2
    n2 = nth
    count += 1
```

```python
# DIY 2: Another method to write a program to find the fibonacci series using while loop.

n = int(input("Enter the number of terms: "))
a = 0
b = 1
i = 1
while i <= n:
    print(a)
    c = a + b
    a = b
    b = c
    i += 1
```

```python
# DIY 3: Write a program to print the values of the fibonacci series upto a certain limit

limit = int(input("Enter the limit: "))
a = 0
b = 1
while a <= limit:
  print(a, end=" ")
  (a, b) = (b, a + b)
```

```python
# DIY 4: Write a program using while loop to reverse the number.
# For num = 0, the loop condition (num > 0) is not entered, so reversed_num remains 0.
# For negative numbers, the loop condition (num > 0) is not entered, so reversed_num remains 0.

num = int(input("Enter a number to reverse: "))

reversed_num = 0
original_num = num

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num = num // 10

print("The original number is: ", original_num)
print("The reversed number is: ", reversed_num)

# Here is how the code works:
# We get the last digit of the number. Then we add the digit to the reversed number. Then we remove the last digit. The final number will be reversed.
```

```python
# DIY 5 Part 1: Write a program to check palindrome for number

num = int(input("Enter a number to check palindrome: "))

temp = num
reversed_num = 0

while temp > 0:
    digit = temp % 10
    reversed_num = reversed_num * 10 + digit
    temp = temp // 10

if num == reversed_num:
    print("The given number is a palindrome number.")
else:
    print("The given number is not a palindrome number.")
```

```python
# DIY 5 Part 2: Write a program to check palindrome for a string.

input_string = input("Enter a string to check palindrome: ")

reversed_string = " "
index = len(input_string) - 1

while index >= 0:
    reversed_string += input_string[index]
    index -= 1

if input_string == reversed_string:
    print("The given string is a palindrome string.")
else:
    print("The given string is not a palindrome string.")
```

```python
# DIY 6: Write a program to count the number of digits in a number.

num = int(input("Enter a number: "))
count = 0

while num > 0:
  count += 1
  num = num // 10
print("Number of digits = ", count)
```

```python
# DIY 6: Write a program to search an element in a list.
nums = [10, 20, 30, 40, 50]
key = int(input("Enter the element to search: "))

i = 0
while i < len(nums):
    if nums[i] == key:
        print("Element found at index", i)
        break
    i += 1
else:
    print("Element not found in the list")
```
