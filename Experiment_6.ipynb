```python
# DIY Problem 2: Write a python program to increment dates

date=input("Enter date in dd/mm/yyyy format: ")
dds,mms,yys=date.split("/")
dd = int(dds)
mm = int(mms)
yy = int(yys)
if(mm==1 or mm==3 or mm==5 or mm==7 or mm==8 or mm==10 or mm==12):
  max1=31
elif(mm==4 or mm==6 or mm==9 or mm==11):
  max1=30
elif(yy%4==0 and yy%100!=0 or yy%400==0):
  max1=29
else:
  max1=28
if(mm<1 or mm>12):
  print("Date is invalid")
elif(dd<1 or dd>max1):
  print("Date is invalid")
elif(dd==max1 and mm!=12):
  dd=1
  mm=mm+1
  print("The incremented date is: ",dd,mm,yy)
elif(dd==31 and mm==12):
  dd=1
  mm=1
  yy=yy+1
  print("The incremented date is: ",dd,mm,yy)
else:
  dd=dd+1
  print("The incremented date is: ",dd,mm,yy)
```

```python
# DIY Problem 3: Write a Python program to calculate the gross salary of an employee based
# If basic <= 10000: HRA = 20% , DA = 80%
# If basic <=20000: HRA = 25% , DA = 90%
# Otherwise : HRA = 30% , DA = 95%

basic = int(input("Enter salary: "))
if basic <=10000:
  hra = basic*0.20
  da = basic*0.80
elif basic<=20000:
  hra = basic*0.25
  da = basic*0.90
else:
  hra = basic*0.30
  da = basic *0.95

gross = basic + hra + da
print("Gross Salary: ", gross)
```

```python
# DIY Problem 4: Write a Python program to calculate income tax based on annual income salary
# Upto 250000 : No Tax
# 250001 - 500000 : 5%
# 500001 - 1000000 :20%
# Above 1000000 : 30%

sal = int (input("Enter salary: "))
if sal<=250000:
  tax = 0
elif sal<=500000:
  tax = (sal - 250000)*0.05
elif sal<=1000000:
  tax = (sal-500000)*0.20 + 250000*0.05
else:
  tax = (sal-1000000)*0.30 + 500000*0.20 + 250000*0.05

print("Total Income Tax: ",tax)
```
